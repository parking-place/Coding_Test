```python
sum = 8
```

```python
test_nums = [-7, -3, -2, 5, 8]
```

```python
def nums_to_funcs(nums):
    funcs = []
    for num in nums:
        f = lambda x: 1 + x**num
        funcs.append(f)
    return funcs
```

```python
test_nums_funcs = nums_to_funcs(test_nums)
```

```python
def get_one_n_sqrts(n):
    import cmath
    n_sqrts = []
    if n == 0:
        return [0.1+1j]
    for i in range(0, n):
        n_sqrt = cmath.rect(1, 2 * math.pi * i / n)
        n_sqrts.append(n_sqrt)
    return n_sqrts
```

```python
one_sqrts = get_one_n_sqrts(sum)
```

```python
for rt in one_sqrts:
    results = []
    result = 1
    for func in test_nums_funcs:
        result *= func(rt)
    results.append(result)

results_sum = 0
for result in results:
    results_sum += result

print(results_sum)
```

```python
def get_one_n_sqrts_polar(n):
    import cmath
    n_sqrts = []
    if n == 0:
        return [1+1j]
    for i in range(0, n):
        n_sqrt = (1, i)
        n_sqrts.append(n_sqrt)
    return n_sqrts

def get_one_n_sqrts(n):
    import cmath
    n_sqrts = []
    if n == 0:
        return [1+1j]
    for i in range(0, n):
        n_sqrt = cmath.rect(1, 2 * math.pi * i / n)
        n_sqrts.append(n_sqrt)
    return n_sqrts
```
